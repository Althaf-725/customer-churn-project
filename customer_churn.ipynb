# REMOVING ALL DEUPLICATE AND NULL VALUES

```python
Data.isnull().sum()
```

```python
Data=Data.drop_duplicates()
```

```python
Data=Data.dropna()
```

```python
Data
```

# FEATURE SCALING

```python
Data.dtypes
```

```python
Data['TotalCharges'] = pd.to_numeric(Data['TotalCharges'], errors='coerce')
```

#FEATURE SCALING

```python
import numpy as np 
a = Data.select_dtypes(include=np.number)
```

```python
a
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
a=StandardScaler().fit_transform(a)
a
```

```python
b=pd.DataFrame(a,columns=["	SeniorCitizen","tenure","Monthly_Charges","Total_Charges"])
```

```python
b
```

# ENCODING

```python
c= Data.select_dtypes(include=["object"])
```

```python
c
```

```python
d=c.drop("customerID",axis=1)
```

```python
d
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in d:
    d[i]=LabelEncoder().fit_transform(d[i])
```

```python
d.head()
```

# COMBINING THE NUMERICAL & CATEGORICAL COLUMNS

```python
Data1=pd.concat([b,d],axis=1)
```

```python
Data1=Data1.dropna()
```

# DEIFING THE TARGET AND FEATURE

```python
y=Data1["Churn"]
x=Data1.drop("Churn",axis=1)
```

```python
x.head()
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# MODEL DEPLOYMENT

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

# PRESICTED VALUES

```python
y_pred=model.predict(x_test)
```

# METRICES

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
import warnings
warnings.filterwarnings("ignore")
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
